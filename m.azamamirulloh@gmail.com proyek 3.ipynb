Dikatakan salah jika kondisinya mobil tua yang disimpan lama digarasi atau mobil antik yang jarang dipakai. Namun kita akan mengabaikan hal tersebut, untuk membuktikan **model_year** memiliki hubungan yang kuat dengan **odometer** maka digunakan tabel corr.

Terbukti bahwa **model_year** dan **odometer** memiliki korelasi yang paling tinggi yaitu sebesar -0.46 (Moderate Negative). Yang artinya semakin kecil nilai model year (makin tua) maka odometernya akan semakin besar. Hal tersebut sejalan dengan pernyataan sebelumnya.

Selanjutnya perlu mempelajari sebaran data di kolom **odometer**

````python
df["odometer"].hist(bins=100)
````

````python
df.describe().loc[["25%","50%","75%"],"odometer"]
````

Dari grafik histogram diatas dapat dilihat bahwa sebaran data di kolom **odometer** agak skewed dengan adanya outliers. Sehingga pengisian data yang hilang lebih baik menggunakan data median bukan mean.

Strategi selanjutnya kita akan mengecek median dan rerata dari masing-masing kelompok usia mobil. Pertama kita buat fungsi untuk pengelompokan usia. Dimana nantinya nilai kosong akan diisi oleh median dari kelompok usia mobil.

````python
df["model_year"].hist(bins = 100)
````

Dapat dilihat bahwa histogram kolom **model_year** skewed ke kanan yang artinya sebagian besar data berart pada nilai 2000 - 2020. Karena tidak terdistribusi nomral maka pengelompokkan range usia didasari dengan nilai persen sebaran data.

````python
persen_data = np.linspace(0.1,0.8,8)
model_year_persen = list(df["model_year"].quantile(persen_data))
model_year_persen
````

Dari data diatas, ketentuan pengelompokan usia:
- **model_year** <= 2003 = kelompok 1
- 2003 < **model_year** <= 2005 = kelompok 2
- ...
- 2015 < **model_year** = kelompok 9

Selanjutnya yaitu membuat fungsi untuk pengelompokan usia.

````python
def pengelompokanUsia(data):
    usia = data["model_year"]
    if usia>model_year_persen[-1]:
        return 9
    if usia<model_year_persen[0]:
        return 1
    for i in range(len(model_year_persen)):
        if model_year_persen[i] <= usia <= model_year_persen[i+1]:
            return i+1
````

Menguji function

````python
pengelompokanUsia(df.loc[1,:])
````

Fungsi berjalan dengan baik. Selanjutnya menerapkan fungsi ke dalam dataframe.

````python
df["age_group"] = df.apply(pengelompokanUsia,axis=1)
df.head()
````

### Menambahkan usia kendaraan saat iklan ditayangkan

````python
# Tambahkan usia kendaraan saat iklan ditayangkan
def usiaMobil(data):
    tahun_rilis = data['model_year']
    tahun_iklan = data['year_posted']
    if tahun_rilis == tahun_iklan:
        return 1
    return tahun_iklan - tahun_rilis
````

## Mempelajari dan Menangani Outlier



Terdapat tiga kolom yang memiliki outlier sesuai dengan analisis di atas yaitu **price, odometer, dan car_age**

````python
def dataRes(data,col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3-Q1
    mini = Q1-1.5*IQR
    maxi = Q3+1.5*IQR
    if mini<0: mini = 0
    return mini, maxi #batas bawah, batas atas
````

### Kolom Harga 

````python
min_price, max_price = dataRes(df,"price")
df_no_outlier = df.query('@min_price <= price <= @max_price')
````

Mari membuat pair plot (Just for fun!)

````python
def PP(data,col):
    data = data[col]
    panjang = len(col)
    k = 0
    plt.figure(0,figsize=(16,16))
    for i in range(panjang):
        for j in range(panjang):
            k+=1
            plt.subplot(panjang,panjang,k)
            sumbux = data[col[j]]
            sumbuy = data[col[i]]
            plt.plot(sumbux,sumbuy,'.')
            plt.plot(np.unique(sumbux), np.poly1d(np.polyfit(sumbux, sumbuy, 1))
                     (np.unique(sumbux)), color='red')
            if j==0:
                plt.ylabel(col[i])
            if i==len(col)-1:
                plt.xlabel(col[j])
            plt.grid()
    plt.show()
````

````python
PP(df_no_outlier,['price','odometer','car_age','condition_value'])
````

Grafik pair plot di atas menunjukan korelasi antar kolom dengan bantuan garis merah yang merupakan garis interpolasi antar dua kolom. Dari garis merah tersebut kita bisa tahu hubungan antar kolom, jika garis merah meningkat maka hubungannya positive sebaliknya jika garis merah menurun maka hubungannya negative.
